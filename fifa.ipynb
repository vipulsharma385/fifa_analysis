Maximum players are from England

```python
df['nationality'].value_counts()[0:10]
```

```python
plt.figure(figsize=(10,10))
col=['red','gray','pink','blue','lightblue','yellow','green','purple','darkred','orange']
plt.bar(list(df['nationality'].value_counts()[0:10].keys()),list(df['nationality'].value_counts()[0:10]),color=col)
plt.title('Top 10 Countries with Maximum Number of Players')
plt.show()
```

```python
ply_sal=df[['short_name','wage_eur']]
```

```python
ply_sal.head()
```

```python
ply_sal=ply_sal.sort_values(by=['wage_eur'],ascending=False)
```

```python
plt.figure(figsize=(5,3))
plt.barh(list(ply_sal['short_name'][0:5]),list(ply_sal['wage_eur'][0:5]),color='hotpink')
plt.title('Top 10 Players with Maximum Salaries')
```

Lionel Messi gets the highest wages(euros) out of all the fifa playing players

```python
#Now checking wages for japan players
jpn=df[df['nationality']=='Japan']
jpn=jpn.sort_values(by=['wage_eur'],ascending=False)
```
